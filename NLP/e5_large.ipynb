```python
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine
model = SentenceTransformer('intfloat/multilingual-e5-large')

def inference(sent1, sent2):
  cosine_dist = cosine(model.encode(sent1), model.encode(sent2))
  if cosine_dist < 0.01:
    return 5
  elif cosine_dist < 0.034:
    return 4
  elif cosine_dist < 0.047:
    return 3
  elif cosine_dist < 0.06:
    return 2
  else:
    return 1
```
